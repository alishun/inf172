```python
import pandas as pd
from kmodes.kprototypes import KPrototypes
```

```python
df = pd.read_csv('dataset.csv')
df_strings = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']

# Calculate BMI and drop height, weight columns
df['BMI'] = df['Weight'] / (df['Height'] ** 2)

df.drop('Height', axis=1, inplace=True)
df.drop('Weight', axis=1, inplace=True)
```

```python
df
```

```python
X = df.drop(columns = ['NObeyesdad'])
X
```

```python
kproto = KPrototypes(n_clusters=7, init='Huang', verbose=2)

# Fit the model
clusters = kproto.fit_predict(X, categorical=[0, 2,3,6,7,9,12,13])

# Assign the cluster labels to the original DataFrame
X['Cluster'] = clusters
```

```python
kproto.cluster_centroids_
```

```python
centroids = kproto.cluster_centroids_
center_df = pd.DataFrame(centroids, columns=X.columns[:-1])  # Exclude the 'Cluster' column
#center_df['Cluster'] = center_df['Cluster'].round().astype(int)

print(center_df)
```
